```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criar DataFrame com os dados dos dias e número de clientes
dados_dias = pd.DataFrame({
    'dia_semana': [1, 2, 3, 4, 5, 6, 7],
    'nº_de_clientes': [15, 20, 30, 100, 350, 500, 700]
})
```

```python
# Aplicar o algoritmo KMeans
X = dados_dias.values
```

```python
kmeans = KMeans(n_clusters=3)  # Definindo 3 clusters
kmeans.fit(X)
```

```python
# Obter os centroides e rótulos dos clusters
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python

# Plotar o gráfico das visualizações
plt.scatter(dados_dias['dia_semana'], dados_dias['nº_de_clientes'], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', color='red', s=200)  # Plotar os centroides
plt.xlabel('Dia da semana')
plt.ylabel('Número de clientes')
plt.title('Agrupamento dos Dias de Maior Movimento')
plt.show()
```
