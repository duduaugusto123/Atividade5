```python
#Passo 1 importação das bibliotecas
import numpy as np #Biblioteca para criar o array numpy
import pandas as pd #Biblioteca para criar o dataframe
from sklearn.cluster import KMeans #Bibioteca para plotagem de graficos
import matplotlib.pyplot as plt #Biblioteca Nova
```

```python
# Passo 2 Criar um DataFrame com os dados dos setores
dados_setores = pd.DataFrame({
 'Setores': [1,2,3,4,5,6,7,8],
 'num de produtos fabricados': [100,50,15,200,500,1000,375,450]
})
```

```python
dados_setores.head
```

```python
# Aplicar o algoritmo KMeans
X = dados_setores.values  # Definir os dados de treinamento
```

```python

kmeans = KMeans(n_clusters=3)  # Definindo 3 clusters
kmeans.fit(X)  # Treinar o modelo
```

```python
# Avaliar o desempenho do algoritmo KMeans
sse = kmeans.inertia_
print('SSE: ', sse)
```

```python
# Obter os centroides e rótulos dos clusters
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
# Plotar o gráfico das visualizações
plt.scatter(dados_setores['Setores'], dados_setores['num de produtos fabricados'], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', color='red', s=200)  # Plotar os centroides
plt.xlabel('Setores')
plt.ylabel('Número de produtos fabricados')
plt.title('Agrupamento de Setores por Produtividade')
plt.show()
```
