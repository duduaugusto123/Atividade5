```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Criar DataFrame com os dados das máquinas
dados_maquinas = pd.DataFrame({
    'Máquina': ['M1', 'M2', 'M3', 'M4', 'M5', 'M6', 'M7', 'M8', 'M9', 'M10'],
    'Temperatura (°C)': [70.2, 65.1, 75.5, 80.3, 68.7, 72.9, 78.6, 66.4, 73.1, 69.5],
    'Vibração': [12.5, 8.2, 15.6, 10.2, 11.8, 14.3, 9.8, 8.9, 13.7, 12.1],
    'Corrente': [4.7, 3.9, 5.1, 4.5, 4.2, 5.3, 4.8, 4.0, 5.0, 4.3]
})
```

```python
# Criar matriz numpy com os dados
X = dados_maquinas.iloc[:, 1:].values
```

```python
# Aplicar o algoritmo KMeans
kmeans = KMeans(n_clusters=3)  # Definindo 3 clusters
kmeans.fit(X)
```

```python
# Obter rótulos dos clusters
labels = kmeans.labels_
```

```python
# Plotar scatter plot dos clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Vibração')
plt.title('Clusters das máquinas industriais')
plt.colorbar(label='Clusters')
plt.show()
```
