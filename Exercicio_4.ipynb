```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criar DataFrame com os dados das substâncias
dados_substancias = pd.DataFrame({
    'Substância': ['Álcool', 'Gasolina', 'Leite', 'Querosene', 'Óleo', 'Vinho'],
    'Concentração (%)': [12.5, 0.1, 4.0, 1.2, 0.5, 15.0],
    'Teor Alcoólico (%)': [50, 0.05, 0.01, 0.02, 0.01, 12.5]
})
```

```python
# Exibir os dados
print(dados_substancias)
```

```python
# Criar matriz de dados
X = dados_substancias[['Concentração (%)', 'Teor Alcoólico (%)']].values
```

```python
# Aplicar o algoritmo KMeans
kmeans = KMeans(n_clusters=3)  # Definindo 3 clusters
kmeans.fit(X)
```

```python
# Obter os centroides e rótulos dos clusters
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
# Plotar o gráfico das visualizações
plt.scatter(dados_substancias['Concentração (%)'], dados_substancias['Teor Alcoólico (%)'], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', color='red', s=200)  # Plotar os centroides
plt.xlabel('Concentração (%)')
plt.ylabel('Teor Alcoólico (%)')
plt.title('Agrupamento de Substâncias')
plt.show()
```
